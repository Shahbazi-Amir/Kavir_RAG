## RAG Flow (English, )

**START**

1. **Input File** (PDF / TXT / MD)
2. **Ingestion**  extract text (no full RAM load)
3. **Chunking**  split text into overlapping chunks
4. **Embedding**  each chunk  vector
5. **Vector DB**  store vectors + metadata

**QUERY TIME**
6) **User Question**
7) **Embedding (query)**
8) **Retrieval (Top-K chunks)**
9) **Prompt = Question + Retrieved Context**
10) **Local LLM  Answer**

**END**
